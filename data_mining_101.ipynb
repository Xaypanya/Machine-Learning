```python
import pandas as pd
import numpy as np
df = pd.read_excel('/content/sample_data/score1.xlsx')
# 1.ສະແດງຂໍ້ມູນໃນຕາຕະລາງທັງໝົດ
df
```

```python
# 2.ໃນຕາຕະລາງມີຈັກຖັນ ແລະ ຈັກແຖວ
df.shape
```

```python
# 3.ໃຫ້ສະແດງຊື່ຖັນ
list(df.columns)
```

```python
# 4.ປ່ຽນເພດຈາກ ພຣະ, ສນ, ທ້າວ ໃຫ້ເປັນ M ແລະ ນາງ ເປັນ F
df['gender'] = df['gender'].replace({'ພຣະ':'M', 'ສນ':'M', 'ທ້າວ':'M', 'ນາງ':'F'})
```

```python
# 5.ຊອກຫາຖັນທີ່ມີຄ່າຫວ່າງ
df.isnull().sum(axis=0)
```

```python
# 6.ແທນຄ່າຫວ່າງດ້ວຍ 0
df = df.fillna(0)
```

```python
# 7.ເພີ່ມຖັນ Total ເຂົ້າ DataFrame ເຊິ່ງຄ່າໄດ້ມາຈາກ ການເອົາ 3 ຖັນບວກກັນ
df['Total'] = df[['Attendent', 'Lecture', 'Lab']].sum(axis=1)
```

```python
# 8.ສະແດງຂໍ້ມູນໃນຕາຕະລາງທັງໝົດ
df
```

```python
# 9.ໃນຕາຕະລາງມີຈັກຖັນ ແລະ ຈັກແຖວ
df.shape
```

```python
# 10.ເພິ່ມຖັນ Grade ເຊິ່ງຄິດໄລ່ມາຈາກ Total(ກຳນົດເງື່ອນໄຂ ຖ້າຄະແນນຫຼາຍກວ່າ 90 ສະແດງຂໍ້ຄວາມ Excellent ຄະແນນຫຼາຍກວ່າ 70 ແມ່ນ Better ຄະແນນຫຼາຍກວ່າ 50 Good ແລະ ຕ່ຳກວ່າ 50 Fail)
df['Grade'] = np.select([df['Total']>=90, df['Total']>=70, df['Total']>=50, df['Total']<50], ['Excellent', 'Better', 'Good', 'Fail'])
```

```python
# 11.ສະແດງລາຍຊື່ນັກສຶກສາທີ່ໄດ້ Excellent
df[df['Grade'] == "Excellent"]['Name']
```

```python
# 12.ສະແດງລາຍຊືີ່ນັກສຶກສາເພດຍິງ ທີ່ໄດ້ຄະແນນ Lecture ຫຼາຍກວ່າ 10 ຂຶ້ນໄປ
df[df['Lecture'] > 10][df['gender'] == "F"]['Name']
```
